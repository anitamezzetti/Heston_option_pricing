```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def stock_price (n, S0, V0, mu, kappa, theta, sigma, dt):
    # W1 = np.random.rand(n)
    # W2 = np.random.rand(n)
    
    V = V0
    S = S0
    stock_path = []
    
    for i in range(n):
        
        W1 = np.random.rand(1)
        W2 = np.random.rand(1)
    
        V = V + kappa * (theta-V) * dt + sigma * np.sqrt(V) * W2
        S = S + mu * S * dt + S * sigma * np.sqrt(V) * W1
        stock_path.append(float(S))
        
    return stock_path
        
```

```python
# parameters 

theta = 0.04
kappa = 1.5
sigma = 0.3
mu = -0.6
V0 = 0.0411
S0 = 1202.10

T =1

```

```python
n = 1000

path = stock_price (n, S0, V0, mu, kappa, theta, sigma, dt)
```
